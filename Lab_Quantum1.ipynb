```python
# Implement a half adder in quantum circuit: // other example of circuit
circ = QuantumCircuit(4, 2)

# Apply X-gate to the first qubit to represent input A
circ.x(0)

# Apply X-gate to the second qubit to represent input B
circ.x(1)

# Use a Toffoli gate to perform the half adder operation
circ.ccx(0, 1, 2)
circ.cx(0, 1)

# Measure the output qubits to obtain the sum and carry bits
circ.measure(1, 0)
circ.measure(2, 1)


result1 = job.result()  # get the results
# interpret the results as a "counts" dictionary
result1.get_counts()

# Draw the circuit
circ.draw()
```
